```python
import folium
```

```python
df=pd.read_csv('substations_all.csv')
```

```python
df.columns= df.columns.str.strip()
```

```python
df.head()
```

```python
subset_of_df = df.sample(n=500)
```

```python
some_map = folium.Map(location=[subset_of_df['Y'].mean(),subset_of_df['X'].mean()],zoom_start=10, control_scale=True)
```

```python
for row in subset_of_df.itertuples():
    some_map.add_child(folium.Marker(location=[row.Y, row.X], popup=row.Name))
```
